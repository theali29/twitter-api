```python
import tweepy as ts
my_api_key="2PemVZGwVJAfyhRZdVsSFxjJk"
my_api_secret="SKGsCTNLgNIGU2CZPToqZg30Bw1segizYAinuOl5VgirB678j5"
```

```python
auth=ts.OAuthHandler(my_api_key,my_api_secret)
api=ts.API(auth,wait_on_rate_limit=True)
```

```python
search_query="#foundropwor -filter:tweets" #tweeted with this hashtag using my account
#fetching tweets with the hashtag foundropwor
```

```python
tweets=ts.Cursor(api.search_tweets,
                 q=search_query,
                 lang="en",
                 since="2020-08-04").items(100)
tweets_copy=[]
for x in tweets:
    tweets_copy.append(x)
print("Total Tweets found:",tweets_copy)
```

```python
import pandas as pd
tweets_df=pd.DataFrame()
for tweet in tweets_copy:
    hashtags=[]
    try:
        for hashtag in tweet.entities["hashtags"]:
            hashtags.append(hashtag["text"])
        text=api.get_status(id=tweet.id,tweet_mode='extended').full_text
    except:
        pass
    tweets_df=tweets_df.append(pd.DataFrame({'user_name':tweet.user.name,
                                             'user_location':tweet.user.location,
                                             'user_description':tweet.user.description,
                                             'user_verified':tweet.user.verified,
                                             'date':tweet.created_at,
                                             'tweet':text,
                                             'hashtags':[hashtags if hashtags else None],
                                             'source':tweet.source}))
    tweets_df=tweets_df.reset_index(drop=True)

tweets_df.head(50)
```
